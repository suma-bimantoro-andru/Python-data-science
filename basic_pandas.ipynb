```python
data[
    (data['country']=='indonesia') &
    (data['year']>2015) &
    (data['genre']=='indonesia pop')
].sort_values('dur', ascending=False)
```

F-Group By

```python
#Mencari nilai rata- rata durasi lagi per country
```

```python
data_duration=data.groupby('country')['dur'].mean().sort_values()
data_duration
```

```python
#Mengubah ke dataframe
data_duration=data_duration.to_frame().reset_index()
data_duration.columns=['country','duration']
data_duration
```

```python
#Plot duration
data_duration.plot(x='country',
                  y='dur', 
                  title='tingkat durasi rata - rata',
                   figsize=(15,5),
                   kind='bar')
```

save dataset

```python
data_duration.to_csv('data_duration.csv', index=False)
```
